```python
%matplotlib inline
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
import dask.array as da
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.ticker as ticker
import cmocean.cm as cmo
import h5py
from datetime import datetime as dt
from scipy.interpolate import interp1d

plt.style.use('ggplot')
```

```python
dataurl = 'http://oceanmodeling.pmc.ucsc.edu:8080/thredds/dodsC/ccsra_2016a_phys_agg_zlevs/fmrc/CCSRA_2016a_Phys_ROMS_z-level_(depth)_Aggregation_best.ncd'
```

```python
### Open the dataset with xarray
ds = xr.open_dataset(dataurl)
```

```python
### Let's see what's in the dataset
list(ds)
```

```python
### making a timenum column for plotting and sigma zed coordinates
ds['mtime'] = mdates.date2num(ds.time)
ds['sigma'] = ds.z / np.nanmax(np.abs(ds.z.values)) # we can do this because there is no coordinate stretching
```

```python
### Import modules for mapping
import cartopy.crs as ccrs
import cartopy
import matplotlib.gridspec as gridspec # GRIDSPEC !
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
```

```python
time_index = 1800
var = ds.rho[time_index,4,:,:]
```

```python
fig = plt.figure(1, figsize=(13,8))
ax = plt.subplot(projection=ccrs.PlateCarree())
plot_date = mdates.num2date(ds.mtime.values[time_index])
plt.title('density ({}$m$) \n CaliROMS on {}'.format(ds.z.values[4],plot_date))

max_v = np.ceil(np.nanmax(var))
min_v = np.floor(np.nanmin(var))
kw = dict(levels=np.linspace(min_v,max_v,101), 
          cmap=cmo.dense, 
          transform=ccrs.PlateCarree())

plt.contourf(ds.lon_rho.values, ds.lat_rho.values, var, **kw)
cb = plt.colorbar(pad=0.01)
cb.set_label('$\sigma$-density ($kg$ $m^{-3}$)',rotation=270, labelpad=25, fontsize=12)

### High resolution coastline is set by resolution='10m'
ax.coastlines(resolution='50m', color='black', linewidth=1)
ax.add_feature(cartopy.feature.LAND, zorder=0)
ax.add_feature(cartopy.feature.OCEAN, zorder=0, color='gray', alpha=0.1)

gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                  linewidth=1, color='gray', alpha=0.2, linestyle='--')
gl.xformatter = LONGITUDE_FORMATTER
gl.yformatter = LATITUDE_FORMATTER
gl.xlabels_top = False
gl.ylabels_right = False
gl.xlabel_style = {'size': 14}
gl.ylabel_style = {'size': 14}

plt.show()
```

```python
## fill the weird surface nans?
rho = ds.rho[time_index,:,:,:] + 1000
rho = rho.values

zlen, xlen, ylen = rho.shape        
```

```python
for i in range(xlen):
    for j in range(ylen):
        q = np.where(np.isnan(rho[:,i,j]))[0]
        if len(q) > 0:
            if len(q) < 5:
                qminus1 = np.nanmin(q)-1
                rho[q,i,j] = rho[qminus1,i,j]
```

```python
var = rho[10,:,:]
fig = plt.figure(1, figsize=(13,8))
ax = plt.subplot(projection=ccrs.PlateCarree())
plot_date = mdates.num2date(ds.mtime.values[time_index])
plt.title('density ({}$m$) \n CaliROMS on {}'.format(ds.z.values[10],plot_date))

max_v = np.ceil(np.nanmax(var))
min_v = np.floor(np.nanmin(var))
kw = dict(levels=np.linspace(min_v,max_v,101), 
          cmap=cmo.dense, 
          transform=ccrs.PlateCarree())

plt.contourf(ds.lon_rho.values, ds.lat_rho.values, var, **kw)
cb = plt.colorbar(pad=0.01)
cb.set_label('$\sigma$-density ($kg$ $m^{-3}$)',rotation=270, labelpad=25, fontsize=12)

### High resolution coastline is set by resolution='10m'
ax.coastlines(resolution='50m', color='black', linewidth=1)
ax.add_feature(cartopy.feature.LAND, zorder=0)
ax.add_feature(cartopy.feature.OCEAN, zorder=0, color='gray', alpha=0.1)

gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                  linewidth=1, color='gray', alpha=0.2, linestyle='--')
gl.xformatter = LONGITUDE_FORMATTER
gl.yformatter = LATITUDE_FORMATTER
gl.xlabels_top = False
gl.ylabels_right = False
gl.xlabel_style = {'size': 14}
gl.ylabel_style = {'size': 14}

plt.show()
```

```python
## Calculate the PEA using the rough grid here
## Note this is extra easy because this ROMS grid has a very lazy vertical grid structure
sigma = ds.sigma.values
d_len  = len(sigma)
rho_0 = np.nanmean(rho)

density = rho[:,:,:] - rho_0
a = np.where(np.isnan(density[d_len-1,:,:]))
empty   = np.where(np.isnan(density))
rhomean = np.trapz(density,sigma,axis=0)
drho    = (rhomean-density)/(rhomean+rho_0)
torque  = np.ones_like(drho)
for i, dep in enumerate(sigma):
    torque[i,:,:] = drho[i,:,:] * float(dep)
torque[empty] = 0
sea    = np.trapz(torque,sigma,axis=0)
sea[a] = np.nan    
```

```python
var = sea*1000
fig = plt.figure(1, figsize=(13,8))
ax = plt.subplot(projection=ccrs.PlateCarree())
plot_date = mdates.num2date(ds.mtime.values[time_index])
plt.title('Simpson Energy Anomaly \n CaliROMS on {}'.format(plot_date))

max_v = np.nanmax(var)
min_v = np.nanmin(var)
kw = dict(levels=np.linspace(min_v,max_v,101), 
          cmap=cmo.matter_r, 
          transform=ccrs.PlateCarree())

plt.contourf(ds.lon_rho.values, ds.lat_rho.values, var, **kw)
cb = plt.colorbar(pad=0.01)
cb.set_label('1000 x potential energy anomaly / weight of the water column',rotation=270, labelpad=25, fontsize=12)

### High resolution coastline is set by resolution='10m'
ax.coastlines(resolution='50m', color='black', linewidth=1)
ax.add_feature(cartopy.feature.LAND, zorder=0)
ax.add_feature(cartopy.feature.OCEAN, zorder=0, color='gray', alpha=0.1)

gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                  linewidth=1, color='gray', alpha=0.2, linestyle='--')
gl.xformatter = LONGITUDE_FORMATTER
gl.yformatter = LATITUDE_FORMATTER
gl.xlabels_top = False
gl.ylabels_right = False
gl.xlabel_style = {'size': 14}
gl.ylabel_style = {'size': 14}

plt.show()
```

This is very weird. A negative PEA is a dynamically unstable water column that is being held in place by other physics or my code might just be wrong. But that we can see a lot of the density features from the 80m slice is at least a good sign in some respects.
